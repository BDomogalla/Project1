```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
```

```python
# make path to files (awards)
csv_awards = "AwardsPlayers.csv"
csv_people = "People.csv"
csv_hof = "HallOfFame.csv"

# read file and make into dataframe
awards_df = pd.read_csv(csv_awards,
                             keep_default_na=False, na_values=[""])
hof_df = pd.read_csv(csv_hof,  
                             keep_default_na=False, na_values=[""])
#check data frames
#awards_data_df.head()
#hof_df

#create merged drata frame of hof and awards
merged_df = pd.merge(awards_df, hof_df, on='playerID', how="left")
#rename column titles
merged_df.rename(columns={"yearID_x":"award_year", "yearID_y":"year_inducted"}, inplace=True)

#show only row values under category column to show "player" 
results = merged_df.loc[merged_df.category == "Player",("playerID","awardID", "award_year" , "inducted", "year_inducted", "category")]
#results

#sort to see summary of players who got awards but not necessarily in hof
awardsummary = results.sort_values(['inducted','category', 'playerID','award_year','awardID'])
awardsummary = pd.DataFrame(awardsummary)
awardsummary.head()
```

```python
#lets see how many how many players got which awards INCLUDING both inducted and not inducted using groupby....count()

groupbyawardID = awardsummary.groupby("awardID")["playerID"].count()
allawards = groupbyawardID.to_frame()
#allawards

#list of only inducted HOF players that got awards and orgnize
hofawards0 = results.loc[merged_df.inducted == "Y", ("playerID","awardID", "award_year" , "inducted", "year_inducted", "category")]
#awardsandhof.head()

#lets see how many awards per player using groupby....count()
hofawards0.groupby("awardID")["playerID"].count()

#convert to dataframe
groupby2 = hofawards0.groupby("awardID")["playerID"].count()
hofawards = groupby2.to_frame()
#hofawards 

final = pd.merge(allawards, hofawards, on='awardID', how="left")
final = final.rename(columns={"playerID_x":"all_players", "playerID_y":"HOF"},)
final['HOF_percentage_of_all_players'] = final["HOF"]/final["all_players"]*100

final
    
#See how many HOF awards - 2370 total\n",
#HOF_avg = final['HOF'].sum()\n",
#print(HOF_avg)"
#see how many different awards there are for people who are in the HOF
#awards1 = len(hofawards0['awardID'].unique())
#awards1 "26 different awards"

#awards_df
#see how many different awards there are all players including ones that
#awards = len(awardsummary['awardID'].unique())

#awards "27 different awards"
```

```python
#looking at playerID then awardID

groupbyplayerID = awardsummary.groupby("playerID")["awardID"].count()
allplayers = groupbyplayerID.to_frame()
#allplayers

groupby3 = hofawards0.groupby("playerID")["awardID"].count()
hofawards1 = groupby3.to_frame()
#hofawards1

final1 = pd.merge(allplayers, hofawards1, on='playerID', how="left")
final1 = final1.rename(columns={"awardID_x":"Total_player_awards", "awardID_y":"HOF_awards"})
#final1

#groupby to see how many times each player has been inducted\n",
hof_results = hof_df.groupby('playerID')["inducted"].count()
hof_results = hof_results.to_frame()
#hof_results.head()

#print new dataset
Total1 = pd.merge(final1, hof_results, on='playerID', how="left")
                  
#Total1['HOF_awards'].replace('16','0' , inplace=True) \n",
Total1['Final_awards'] = Total1["Total_player_awards"]/Total1["inducted"]
#rearrange
#Total1 = Total1[['Total_player_awards','Final_awards', 'inducted', 'HOF_awards']]
Total1 
#Total2[HOF_awards].replace(\"Nan\",\"0\")\n",
#Total1(['HOF_awards'].replace['NaN','NO'], inplace=True)\n",

#Total1"
```

```python
HOF_avg_awards= Total1['HOF_awards'].mean()

#HOF_avg_sum_awards = Total1['HOF_#_awards'].max()
#39
#HOF_avg_sum_awards = Total1['HOF_awards'].min()
#1
HOF_avg_sum_awards = Total1['HOF_awards'].sum()
print("The average number of awards a HOF collects is " + str(HOF_avg_awards) + ".")
print(HOF_avg_sum_awards)

```
