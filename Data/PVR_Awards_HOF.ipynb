```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
```

```python
# make path to files (awards)
csv_awards = "AwardsPlayers.csv"
csv_people = "People.csv"
csv_hof = "HallOfFame.csv"

# read file and make into dataframe
awards_df = pd.read_csv(csv_awards,
                             keep_default_na=False, na_values=[""])
hof_df = pd.read_csv(csv_hof,  
                             keep_default_na=False, na_values=[""])
#check data frames
#awards_data_df.head()
#hof_df

#create merged drata frame of hof and awards
merged_df = pd.merge(awards_df, hof_df, on='playerID', how="left")
#merged_df
merged = merged_df.loc[merged_df.category == "Player",("playerID","awardID", "yearID_x" , "inducted", "yearID_y", "category")]
#merged

#summary of players who got awards but not necessarily in hof
awardsummary = merged.sort_values(['inducted','category', 'playerID','yearID_y','awardID'])
awardsummary = pd.DataFrame(awardsummary)
awardsummary
```

```python
#lets see how many awards per player using groupby....count()
awardsummary.groupby("awardID")["playerID"].count()
```

```python
#list of all players that got awards INCLUDING ones who weren't inducted in the HOF and orgnize
awardsnohof = merged_df.loc[merged_df.category == "Player", ("playerID","awardID", "yearID_x" , "inducted", "yearID_y", "category")]

#awardsnohof.head()

#lets see how many awards per player using groupby....count()
awardsnohof.groupby("awardID")["playerID"].count()
```

```python
awardsnohof = awardsnohof.loc[awardsnohof.inducted == "Y", ("playerID", "awardID", "yearID_x", "inducted", "yearID_y", "category")]
awardsnohof.head()
```

```python
# how many total number of players got award(s)
# find unique players by using .unique()
players = len(awardsnohof['playerID'].unique())
awards = len(awardsnohof['awardID'].unique())

print("The number of total players is " + str(players)+ ".")
print("The number of different awards is " + str(awards)+ ".")
```
